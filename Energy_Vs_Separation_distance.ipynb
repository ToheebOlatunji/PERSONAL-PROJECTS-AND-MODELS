```python
# Import necessary libraries
import numpy as np
import pandas as pd

# Given values
sigma = 3.4  # in angstroms (Å)
epsilon = 0.0104  # in eV
delta_epsilon = epsilon / 11  # delta epsilon

# Energy values for i = 0 to 10
Ei_values = [-epsilon + i * delta_epsilon for i in range(11)]

# Lists to store the results
r1_values = []
r2_values = []
x1_values = []
x2_values = []

# Calculate r1 and r2 for each Ei
for Ei in Ei_values:
    a = 4 * epsilon
    b = -4 * epsilon
    c = -Ei
    discriminant = b**2 - 4 * a * c

    # Solve the quadratic equation to get x1 and x2
    x1 = (-b + np.sqrt(discriminant)) / (2 * a)
    x2 = (-b - np.sqrt(discriminant)) / (2 * a)

    # Calculate the separation distances r1 and r2 in angstroms
    r1 = sigma / x1**(1/6)  # r1 corresponding to x1
    r2 = sigma / x2**(1/6)  # r2 corresponding to x2

    # Append the results to the lists
    r1_values.append(r1)
    r2_values.append(r2)
    x1_values.append(x1)
    x2_values.append(x2)

# Create a DataFrame to display the results
df = pd.DataFrame({
    "i": range(11),
    "E_i (Energy) [eV]": Ei_values,
    "x1": x1_values,
    "x2": x2_values,
    "r1 (Distance 1) [Å]": r1_values,
    "r2 (Distance 2) [Å]": r2_values,
    "a": [4 * epsilon] * 11,
    "b": [-4 * epsilon] * 11,
    "c": [-Ei for Ei in Ei_values],
    "Discriminant (delta)": [b**2 - 4 * a * c for a, b, c in zip([4 * epsilon] * 11, [-4 * epsilon] * 11, [-Ei for Ei in Ei_values])]
})

# Display the DataFrame
df
```
