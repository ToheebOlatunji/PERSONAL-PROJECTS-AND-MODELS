```python
import numpy as np
import matplotlib.pyplot as plt

# Phase space data
D_values = np.linspace(1, 20, 100)  # Grain sizes in m
sigma_values = 1 / (1/1e-3 + 1e-3/(D_values * 1e-5))  # Total conductivity formula

# Selected grain sizes for highlighting
highlight_D = [1, 5, 10, 20]
highlight_sigma = [sigma_values[np.searchsorted(D_values, d)] for d in highlight_D]

# Plot
plt.figure(figsize=(10, 6))

# Main curve
plt.plot(D_values, sigma_values*1000, color='navy', linewidth=3, label='Conductivity Evolution')

# Highlight points
plt.scatter(highlight_D, np.array(highlight_sigma)*1000, s=100, c='red', zorder=5, label='Selected Points')

# Labels and title
plt.xlabel('Grain Size (m)', fontsize=18)
plt.ylabel('Total Conductivity (mS/cm)', fontsize=18)
plt.title('Phase Space: Grain Size - Conductivity Relationship', fontsize=22, fontweight='bold')

# Grid and legend
plt.grid(True, linestyle='-', alpha=0.3)
plt.minorticks_on()
plt.grid(True, which='minor', linestyle=':', alpha=0.2)

# Tick parameters for professional appearance
plt.tick_params(axis='both', which='major', direction='in', top=True, right=True, length=8)
plt.tick_params(axis='both', which='minor', direction='in', top=True, right=True, length=4)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
