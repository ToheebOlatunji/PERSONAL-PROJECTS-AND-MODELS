```python
import numpy as np
import matplotlib.pyplot as plt

# Steady-state concentration profile (Eq. 34)
x = np.linspace(0, 1, 100)  # normalized position
c0 = 1.0  # reference concentration
cL = 0.5  # boundary concentration
phi = 0.1 * x  # linear potential gradient

# Concentration profile
c_profile = c0 * (1 + (cL / c0 - 1) * x) * np.exp(-phi)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, c_profile, color='purple', linewidth=3)
plt.xlabel('Normalized Position $x/L$')
plt.ylabel('Normalized Concentration $c/c_0$')
plt.title('Steady-State Proton Concentration Profile')
plt.grid(True)
plt.show()
```
