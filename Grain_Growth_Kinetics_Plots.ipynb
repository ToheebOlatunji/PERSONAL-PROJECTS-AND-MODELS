```python
import numpy as np
import matplotlib.pyplot as plt

# Plot grain size evolution (Eq. 5)
t = np.linspace(0, 100, 1000)  # time in hours
D0 = 1.0  # initial grain size in m
Kg_values = [0.1, 0.5, 1.0, 2.0]  # different growth rates

plt.figure(figsize=(10, 6))
for Kg in Kg_values:
    D = np.sqrt(D0**2 + 2*Kg*t)
    plt.plot(t, D, label=f'Kg = {Kg} m/h')

plt.xlabel('Time (hours)')
plt.ylabel('Average Grain Size (m)')
plt.title('Grain Growth Kinetics - Normal Growth')
plt.legend()
plt.grid(True)
```
