```python
import numpy as np
import matplotlib.pyplot as plt

# Relationship between grain size and total conductivity (Eq. 11)
D_grain = np.logspace(0, 2, 10)  # 1 to 100 m, fewer points for diamond visibility
sigma_bulk = 1e-3  # S/cm
sigma_gb = 1e-5    # S/cm
delta_gb = 1e-3    # m

sigma_total = 1 / (1 / sigma_bulk + delta_gb / (D_grain * sigma_gb))

plt.figure(figsize=(8, 6))
plt.semilogx(D_grain, sigma_total * 1000, 'gD-', markersize=8, linewidth=2, label='Total Conductivity')

plt.xlabel('Average Grain Size (m)', fontsize=14)
plt.ylabel('Total Conductivity (mS/cm)', fontsize=14)
plt.title('Effect of Grain Size on Total Conductivity', fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
